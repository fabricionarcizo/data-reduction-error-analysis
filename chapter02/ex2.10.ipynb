```python
# Imported packages.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson
```

```python
# Factorial recursive function.
def factorial(n):
    return 1 if n == 0 else n * factorial(n - 1)
```

```python
# Poisson distribution function.
def P(x, mean):
    return (mean ** x / factorial(x)) * np.e ** (-mean)
```

```python
# Poisson distribution parameters.
x = range(10)
mean = 1.69
std = mean * 0.5
```

```python
# Calculates the Poisson manually based on Equation 2.11.
# \frac{\mu^x}{x!} e^\mu
pmf = []
for j in x:
    pmf.append(P(j, mean))
pmf = np.asarray(pmf)
print("Manual Poisson distribution:", pmf)
```

```python
# Calculates the Poisson using Scipy and shows the probability mass function.
pmf = poisson.pmf(x, mean)
print("Scipy Poisson distribution:", pmf)
```

```python
# Show the Poisson distribution histogram.
plt.subplot(2, 1, 1)
plt.title("Poisson Distribution")
plt.bar(x, pmf, width=1, edgecolor="k", alpha=0.35, color="k")
plt.xticks([])
plt.ylim([0, 0.5])

# Show the Gaussian distribution histogram.
plt.subplot(2, 1, 2)
plt.title("Gaussian Distribution")
plt.bar(x, norm.pdf(x, mean, std), width=1, edgecolor="k", alpha=0.35, color="k")
plt.ylim([0, 0.5])
```
