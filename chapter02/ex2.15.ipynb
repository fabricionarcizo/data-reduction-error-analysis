```python
# Imported packages.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# There is a solution for this exercise available on http://www.pa.uky.edu/~kwng/phy335/hw/P2_15.pdf [Posted on October 31, 2007].
```

```python
# Dead time
dead = 200 * 10 ** (-9)

# Number of particles
N = 10 ** 6

# Mean number of particles
mean = 1 * N * dead
```

```python
# Assume that a counter has a dead time of 200 ns (200 \times 10^{-9}s) and is exposed to a beam of 1 \times 10^6 particles per second so that the mean number of particles hitting the counter in the 200-ns time slot is \mu=0.2. From the Poisson probability for this process, find the efficiency of the counter, that is, the ratio of the average number of particles counted to the average number that pass through the counter in the 200-ns time period.

# Efficiency: (1 - e^{-\mu}) / \mu
efficiencies = []
efficiencies.append((1 - np.e ** (-mean)) / mean)
print("Efficiency for %.2f: %.4f." % (mean, efficiencies[-1]))
```

```python
# Repeat the calculation for beam rates of 2, 4, 6, 8, and 10 \times 16^6 particles per second, and plot a graph of counter efficiency as a function of beam rate.
beams = [2, 4, 6, 8, 10]
for i in beams:
    mean = i * N * dead
    efficiencies.append((1 - np.e ** (-mean)) / mean)
    print("Efficiency for %.2f: %.4f." % (mean, efficiencies[-1]))
beams.insert(0, 1)
```

```python
plt.plot(beams, efficiencies, linestyle='--', marker='o')
plt.ylim([0, 1])
plt.xlabel("Beam Intensity (/s) $x10^6$")
plt.ylabel("Efficiency")
```
