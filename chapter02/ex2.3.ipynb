```python
# Imported packages.
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
# Factorial recursive function.
def factorial(n):
    return 1 if n == 0 else n * factorial(n - 1)
```

```python
# Binomial distribution function.
def binomial(x, n, p):
    return (factorial(n) / (factorial(x) * factorial(n - x))) * p**x * (1 - p)**(n - x)
```

```python
# Binomial distribution parameters.
n = 6
p = 0.5
x = range(7)
```

```python
# Calculates the binomial manually based on Equation 2.4.
# (n! / (x! (n - x)!)) p^x (1 - p)^{n - x}
pmf = []
for j in x:
    pmf.append(binomial(j, n, p))
pmf = np.asarray(pmf)
print("Manual binomial distribution:", pmf)
```

```python
# Calculates the binomial using Scipy and shows the probability mass function.
rv = binom(n, p)
print("Scipy binomial distribution:", rv.pmf(x))
```

```python
# Calculates the mean and standard deviation.
mean = n * p
std = (n * p * (1 - p)) ** 0.5
print("Binomial distribution mean:", mean)
print("Binomial distribution standard deviation:", std)
```

```python
# Show the binomial distribution histogram.
plt.bar(x, rv.pmf(x), width=1, edgecolor="k", alpha=0.35, color="k")
plt.axvline(mean, color="k", alpha=0.5, linestyle="solid", linewidth=2, label="Mean")
plt.axvline(mean - std, color="k", alpha=0.5, linestyle="dashed", linewidth=2, label="Standard Deviation")
plt.axvline(mean + std, color="k", alpha=0.5, linestyle="dashed", linewidth=2)
plt.legend()
```
